## The Data

We will be working with a famous titanic data set for these exercises. Later on in the Machine Learning section of the course, we will revisit this data, and use it to predict survival rates of passengers. For now, we'll just focus on the visualization of the data with seaborn:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')

titanic = sns.load_dataset('titanic')
titanic.head()
```

# Exercises

**Recreate the plots below using the titanic dataframe.**

**Note: In order to not lose the plot image, make sure you don't code in the cell that is directly above the plot, there is an extra cell above that one which won't overwrite that plot!* **

```python
# Ex.1
#    A SCATTERPLOT with normal distro on spines of FARE v. AGE

sns.jointplot(x='fare',y='age',data=titanic,kind='scatter')
```

```python
# Ex. 2
# HISTOGRAM counting FARES

sns.distplot(titanic['fare'], kde=False, color='red')
```

```python
# Ex. 3
#    BOXPLOT of ages in different flight classes

sns.boxplot(x="class", y="age", data=titanic)
```

```python
# Ex. 4
# SWARMPLOT showing distributions of AGES in each CLASS

sns.swarmplot(x="class", y="age", data=titanic, palette='rainbow')
```

```python
# Ex. 5
# Simple BARPLOT showing COUNT of each SEX aka. countplot

sns.countplot(data=titanic, x='sex')
```

```python
# Ex. 6
# HEATMAP of correlation between variables in titanic dataset. 

sns.heatmap(titanic.corr(numeric_only=True), cmap='coolwarm')
plt.title('Correlation w/in Titanic Data')
```

```python
# Ex. 7
# BARGRAPH of AGE split by SEX

grid = sns.FacetGrid(titanic, col="sex")
grid.map(plt.hist, "age")
```
