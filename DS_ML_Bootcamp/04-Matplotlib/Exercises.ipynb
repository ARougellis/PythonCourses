#### Import Libraries and Set Up Data

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
#           else use `plt.show()`

x = np.arange(0,100)
y = x*2
z = x**2
```

## Exercise 1

1. Create a figure object called fig using plt.figure()
2. Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
3. Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
# 1.
fig = plt.figure(figsize=(5,3))

# 2.
ax = fig.add_axes([0, 0, 1, 1])

# 3.
ax.plot(x, y, 'b')
ax.set_xlabel('X Label') # NOTE the use of set_ to begin methods
ax.set_ylabel('Y Label')
ax.set_title('Title')
```

## Exercise 2
**i. Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
fig = plt.figure(figsize=(5,3))

axes1 = fig.add_axes([0,0,1,1]) # main axes
axes2 = fig.add_axes([0.2,0.5,.2,.2]) # inset axes
```

**ii. Now plot (x,y) on both axes. And call your figure object to show it.**

```python
# Main
axes1.plot(x, y)
axes1.set_xlabel('x1')
axes1.set_ylabel('y1')

# Inside
axes2.plot(x, y)
axes2.set_xlabel('x2')
axes2.set_ylabel('y2')

fig # Show figure object
```

## Exercise 3

1. Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]
2. Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1]) # main axes
axes2 = fig.add_axes([0.2,0.5,.4,.4]) # inside axes

# Main
axes1.plot(x, z)
axes1.set_xlabel('X')
axes1.set_ylabel('Z')
axes1.set_xlim([0, 100])
axes1.set_ylim([1, 10000])

# Inside
axes2.plot(x, y)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_xlim([20, 22])
axes2.set_ylim([30, 50])
axes2.set_title('Zoom')
```

## Exercise 4
