### i.
Figure out how to get the stock data from Jan 1st 2006 to Jan 1st 2016 for each of these banks. Set each bank to be a separate dataframe.

_Tip:_ Use [this documentation page](https://pandas-datareader.readthedocs.io/en/latest/remote_data.html) for hints and instructions (it should just be a matter of replacing certain values. Use google finance as a source, for example:
    
    # Bank of America
    BAC = data.DataReader("BAC", 'google', start, end)

_NOTE:_ Double check the link above as the API for "google" may not alwsy work. Rn i have to use 'stooq' data as google finace data does not provide necessary index.

```python
# Setting start and end times
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
```

```python
# Bank of America
BAC = data.DataReader("BAC", 'stooq', start, end)

# CitiGroup
C = data.DataReader("C", 'stooq', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'stooq', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'stooq', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'stooq', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'stooq', start, end)
```

### ii. 
Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: `tickers`. 

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

### iii.
Use `pd.concat` to concatenate the bank dataframes together to a single data frame called `bank_stocks`. 
- Set the keys argument equal to the tickers list. 
- Note: pay attention to what axis you concatenate on, given all dfs have the same index (being a custom selected rande of dates), we want to concat them onn into columns to get a multi-lvl index as columns

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys=tickers)
bank_stocks.head()
```

### iv.
Set the column name levels to 'Bank Ticker' and 'Stock Info'. Then check the head of the `bank_stocks` df

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']

bank_stocks.head()
```

# II. EDA

Before continuing, check out the documentation on [Multi-Level Indexing](http://pandas.pydata.org/pandas-docs/stable/advanced.html) and [Using `.xs`](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.xs.html) (this well help us grab cross sections of a multi-index df).

### i.
What is the _max Close price for each bank's stock_ throughout the time period?

```python
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

### ii. 
Create a new empty DataFrame called `returns`. This dataframe _will contain the returns for each bank's stock_.   `returns` are typically defined by:

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

### iii.
Use pandas `pct_change()` method on the `Close` column to create a column representing this return value. 
- Create a for loop
    + for each Bank Stock Ticker create this returns column representing `pct_change()`

```python
for tick in tickers:
    returns[tick+ ' Returns'] = bank_stocks[tick]['Close'].pct_change()
    

returns
```

### iv.
Create a `pairplot` using seaborn of the returns dataframe. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('whitegrid')
sns.pairplot(returns[1:])
plt.tight_layout()
```

## 2. Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

### ii.
Plot the _rolling 30 day average against the Close Price_ for Bank Of America's stock for the year 2008

```python
# BOA_08 = bank_stocks.xs(key='BAC', axis=1)['2008-01-01':'2008-12-31']['Close']
BOA_08 = bank_stocks['2008-01-01':'2008-12-31']['BAC']['Close']

# Rolling Average Plot
BOA_08.rolling(window=30).mean().plot(label='30 Day Avg')

# Close Price Plot
BOA_08.plot(label='Close Price')
```
