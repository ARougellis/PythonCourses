# Pandas Built-in Data Visualization

Here we will learn about __pandas built-in capabilities for data visualization__. It's built-off of matplotlib, but it baked into pandas for easier usage. 

### i. Imports

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

### ii. The Data

There are some fake data csv files you can read in as dataframes:

```python
df1 = pd.read_csv('df1', index_col='Date')
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

## 1. Style Sheets

Matplotlib has [style sheets](http://matplotlib.org/gallery.html#style_sheets) you can use to make your plots look a little nicer. These style sheets include `plot_bmh`,`plot_fivethirtyeight`,`plot_ggplot`, etc. They  create a set of style rules that your plots follow and make all your plots have the same look and feel more professional.

**Before `plt.style.use()` your plots look like this:**

```python
df1['A'].hist()
```

**Import & Call the style**:

```python
import matplotlib.pyplot as plt
```

**Now your plots look like this:**

```python
# GGPLOT
#    - based on an R library
#    - simple
plt.style.use('ggplot')
df1['A'].hist()
```

```python
# BMH (or Bayesian Methods for Hackers)
#    - scientific graphing; shows precision of the plot
plt.style.use('bmh')
df1['A'].hist()
```

```python
# DARK BACKGROUND
plt.style.use('dark_background')
df1['A'].hist()
```

```python
# FIVETHIRTYEIGHT
plt.style.use('fivethirtyeight')
df1['A'].hist()
```

```python
plt.style.use('ggplot')
```

Let's _stick with the ggplot style_ and actually show you how to utilize pandas built-in plotting capabilities!

# I. Plot Types

There are several plot types built-in to pandas, most of them statistical plots by nature:

* `df.plot.area`     
* `df.plot.barh` && `df.plot.bar`
* `df.plot.hist`     
* `df.plot.line`     
* `df.plot.scatter`
* `df.plot.box`  
* `df.plot.hexbin`   
* `df.plot.kde`
* `df.plot.density` 
* `df.plot.pie`

**NOTE:** You can also just call `df.plot(kind='hist')` or replace that kind argument with any of the key terms shown in the list above (e.g. `'box'`,`'barh'`, etc..)
___

## 1. Area

```python
df2.plot.area(alpha=0.4)
```

## 2. Barplots

```python
# SIDE-BY-SIDE BARS
df2.plot.bar()
```

```python
# STACKED BARS
df2.plot.bar(stacked=True)
```

## 3. Histograms

```python
df1['A'].plot.hist(bins=50)
```

## 4. Scatter Plots

```python
df1.plot.scatter(x='A',y='B')
```

You can use the param `c=` to color based off another column value
Use cmap to indicate colormap to use. 
For all the colormaps, check out: http://matplotlib.org/users/colormaps.html

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

Or use the param `s=` to indicate _size based off another column_. The `s=` param needs to be an array, not just the name of a column:

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*200)
```

## 5. BoxPlots

**NOTE:** Can also pass a `by=` argument for groupby aggregator. 

```python
df2.plot.box() 
```

## 6. Hexagonal Bin Plot

A _**Bivariate Data** alternative to scatterplot_:

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b', gridsize=25,cmap='Oranges')
```

____

## 7. Kernel Density Estimation plot (KDE)

```python
# Plotting KDE of _single variable_
df2['a'].plot.kde()
```

```python
# Plotting KDE for _multiple variables_
df2.plot.density()
```

As we can see, using pandas built in data viz tools is am easier method of plotting to that of full-on matplotlib, it balances ease of use with control over the figure.
