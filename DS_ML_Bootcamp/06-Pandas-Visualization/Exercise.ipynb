### Data

Use **df3** to replicate the following plots. 

```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

** Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. See if you can figure out how to stretch it in a similar fashion. Remeber back to your matplotlib lecture...**

```python
# EX. 1
# Create a SCATTERPLOT of col 'a' vs 'b'
#      - with red markers that have boarders

df3.plot.scatter(x='a', 
                 y='b', 
                 c=['red'], 
                 edgecolors="black", 
                 figsize=(12,3))
```

```python
# Ex. 2
# Create a histogram of the 'a' column

df3['a'].plot.hist(edgecolor='black')
```

```python
# Ex. 3
# Recreate the HISTOGRAM above... 
#      - using style sheets to set the style to 'ggplot'
#      - w/ more bins

plt.style.use('ggplot')
df3['a'].plot.hist(bins=25, 
                   edgecolor='white', 
                   alpha=0.5)
```

```python
# Ex.4
# Create a BOXPLOT comparing the `a` and `b` columns

df3[['a', 'b']].plot.box()
```

```python
# Ex. 5
# Create a KDE PLOT of the 'd' column

df3['d'].plot.kde()
```

```python
# Ex. 6
# Customize the KDE PLOT above to... 
#      - increase the linewidth 
#      - make the linestyle dashed. 
#            - Note: usually dont dash a kde plot line

df3['d'].plot.kde(lw=5, ls='--')
```

```python
# Ex. 7
# Create an AREA PLOT of all the columns for just the rows up to 30.

df3.head(30).plot.area(alpha=0.5)
```

## Bonus Challenge!

_Notice how the legend in our previous figure overlapped some of actual diagram. Figure out how to **display the legend outside of the plot**_

```python
fig = plt.figure() # <-- i. create figure
df3.head(30).plot.area(alpha=0.5) # <-- ii. Plot
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5)) # <-- iii. Set Legend
plt.show()
```
