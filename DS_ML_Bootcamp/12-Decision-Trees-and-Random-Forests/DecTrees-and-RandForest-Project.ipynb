# <ins>_Decision Tree & Random Forest Project_</ins>

For this project we will be exploring publicly available data from [LendingClub.com](www.lendingclub.com). Lending Club connects people who need money (borrowers) with people who have money (investors). Hopefully, as an investor you would want to invest in people who showed a profile of having a high probability of paying you back. We will try to create a model that will help predict this.

Lending club had a [very interesting year in 2016](https://en.wikipedia.org/wiki/Lending_Club#2016), so let's check out some of their data and keep the context in mind. This data is from before they even went public.

We will use lending data from 2007-2010 and be **trying to classify and predict whether or not the borrower paid back their loan in full**. You can download the data from [here](https://www.lendingclub.com/info/download-data.action) or just use the csv already provided.

Here are what the columns represent:
* `credit.policy`: $1$ if the customer meets the credit underwriting criteria of LendingClub.com; $0$ otherwise.
* `purpose`: The purpose of the loan (takes values "credit_card", "debt_consolidation", "educational", "major_purchase", "small_business", and "all_other").
* `int.rate`: The interest rate of the loan, as a proportion (a rate of 11% would be stored as 0.11). Borrowers judged by LendingClub.com to be more risky are assigned higher interest rates.
* `installment`: The monthly installments owed by the borrower if the loan is funded.
* `log.annual.inc`: The natural log of the self-reported annual income of the borrower.
* `dti`: The debt-to-income ratio of the borrower (amount of debt divided by annual income).
* `fico`: The FICO credit score of the borrower.
* `days.with.cr.line`: The number of days the borrower has had a credit line.
* `revol.bal`: The borrower's revolving balance (amount unpaid at the end of the credit card billing cycle).
* `revol.util`: The borrower's revolving line utilization rate (the amount of the credit line used relative to total credit available).
* `inq.last.6mths`: The borrower's number of inquiries by creditors in the last 6 months.
* `delinq.2yrs`: The number of times the borrower had been 30+ days past due on a payment in the past 2 years.
* `pub.rec`: The borrower's number of derogatory public records (bankruptcy filings, tax liens, or judgments).

# III. Setting up Models

Let's get ready to set up our data for our Random Forest Classification Model!

## i. Categorical Features

If you call the `.info()` method on the dataframe, you will notice that that the `purpose` column is categorical. This means we need to transform it using dummy variables so sklearn will be able to understand them. This can be done with `pd.get_dummies` (the following wasy of dealing with these columns can be expanded to multiple categorical features if necessary).

**9.**  
Create a list of 1 element containing the string 'purpose'. Call this list `cat_feats`.

````python
cat_feats = ['purpose']
cat_feats
````

**10.**  
Now use ```pd.get_dummies(loans,columns=cat_feats,drop_first=True)``` to create a fixed larger dataframe that has new feature columns with dummy variables. Set this dataframe as `final_data`.

````python
# The purpose of this line is to split up the `purpose` col into...
#        ... `loans_df['purpose'].nunique()` number of boolean columns
final_df = pd.get_dummies(loans_df,columns=cat_feats,drop_first=True)
final_df.head()
````

## ii. Train Test Split

Now its time to split our data into a training set and a testing set!

**11.**  
Use sklearn to split your data into a training set and a testing set.

````python
X = final_df.drop('not.fully.paid',axis=1)
y = final_df['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
````

# IV. Decision Tree

## i. Training a Decision Tree Model

Let's start by training a single decision tree first!

**12.**  
Create an instance of `DecisionTreeClassifier()` called `dtree` and fit it to the training data.

````python
# _Creating_ DecisionTree model object
dtree = DecisionTreeClassifier()

# _Training_ the model
dtree.fit(X_train,y_train)
````

## ii. Predictions and Evaluation of Decision Tree

**13.**  
Create...
- predictions from the test set 
- a classification report
- a confusion matrix

````python
# Testing predictions with test data
dtree_predict = dtree.predict(X_test)
````

````python
print("Decision Tree Confusion Matrix:")
print(confusion_matrix(y_test, dtree_predict))
print("")
print("Decision Tree Classification Report:")
print(classification_report(y_test, dtree_predict))
````

# V. Random Forest

## i. Training the Random Forest model

Now its time to train the Random Forest model!

**14.**  
Create an instance of the `RandomForestClassifier` class and fit it to our final training data .

````python
# Creating RFC model object
rfc = RandomForestClassifier()

# Training model
rfc.fit(X_train, y_train)
````

## ii. Predictions and Evaluation of Random forest

Let's predict off the `y_test` values and evaluate our model.

**15.**  
Predict the class of `not.fully.paid` for the `X_test` data.

````python
rfc_predict = rfc.predict(X_test)
````

**16.**  
Now create a classification report from the results. 

````python
print("Random Forest Classification Report:")
print(classification_report(y_test, rfc_predict))
````

**17.**  
Show the Confusion Matrix for the predictions.

````python
print("Random Forest Confusion Matrix:")
print(confusion_matrix(y_test, rfc_predict))
````
