# K Means Clustering Project 

For this project we will attempt to use KMeans Clustering to cluster Universities into to two groups, Private and Public.

## I. Data

<ins>**NOTE**</ins>: the dataset contains labels, but we will NOT use them for the KMeans clustering algorithm, since that is an unsupervised learning algorithm. When using the Kmeans algorithm under normal circumstances, it is because you don't have labels. In this case we will use the labels to try to get an idea of how well the algorithm performed, but you won't usually do this for Kmeans, so the classification report and confusion matrix at the end of this project, don't truly make sense in a real world setting.

## III. K Means Cluster Creation

Now it is time to create the Cluster labels (using KMeans import from SciKit Learn.

**10.**  
Create an instance of a K Means model with 2 clusters.

```python
k_model = KMeans(n_clusters=2)
```

**11.**  
Fit the model to all the data except for the Private label.

```python
k_model.fit(college_df.drop('Private', axis=1))
```

**12.**  
Print the cluster center vectors

```python
k_model.cluster_centers_
```

### i. Evaluation

There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters, keep in mind, you usually won't have this luxury in the real world.

**13.**  
Create a new column for the df called `Cluster`, which is a $1$ for a Private school, and a $0$ for a public school.

```python
# college_df['Cluster'] = college_df['Private'].apply(lambda private: 1 if private=='Yes' else 0)
# Switching labels to match model prediction lableing of clusters
college_df['Cluster'] = college_df['Private'].apply(lambda private: 0 if private=='Yes' else 1)

college_df
```

**14.**  
Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
print('K Means Clustering Confusion Matrix')
print(confusion_matrix(college_df['Cluster'], k_model.labels_))
print('')
print('K Means Clustering Classification Report')
print(classification_report(college_df['Cluster'], k_model.labels_))
```

Not so bad considering the algorithm is purely using the features to cluster the universities into 2 distinct groups.
