# K Means Clustering with Python



### i. Method Used

__K Means Clustering__ is an unsupervised learning (aka takes in unlabeled data) algorithm (which finds patterns in data) that tries to cluster data based on their similarity. In _k means clustering_, we have to specify the number of clusters we want the data to be grouped into. The algorithm randomly assigns each observation to a cluster, and finds the centroid of each cluster. Then, the algorithm iterates through two steps:
1. Calculate centroid of each cluster by finding the mean vector point
2. Reassign data points to the cluster whose centroid is closest 

These two steps are repeated till the within cluster variation cannot be reduced any further. The cluster variation is calculated as the sum of the euclidean distance between the data points and their respective cluster centroids.

### ii. Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### iii. Create Data

```python
from sklearn.datasets import make_blobs
```

```python
# Create Data
data = make_blobs(n_samples=200, n_features=2, 
                           centers=4, cluster_std=1.8,random_state=101)
```

## I. Visualize Data

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

## II. Creating the Clusters

```python
from sklearn.cluster import KMeans
```

```python
# Creating KMean Model object
kmeans = KMeans(n_clusters=4)

# Training KMean Model
kmeans.fit(data[0])
```

```python
# the model finds the cluster center's coordinates
kmeans.cluster_centers_
```

```python
# The model also labeling each element by cluster
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
# KMeans Prediction
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

# Original
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
