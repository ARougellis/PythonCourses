# _Keras Syntax Basics_

With TensorFlow 2.0 versions, Keras is now the main API choice. To understand the basics of the Keras syntax and adding layers, we'll work through a simple regression project.

## I. The Data

To learn the basic syntax of Keras, we will use a very simple fake data set. The data has measurements of some rare gem stones, with 2 measurement features and a sale price. Our final goal would be to try to predict the sale price of a new gem stone we just mined from the ground, in order to try to set a fair price in the market.

### i. Load the Data

```python
import pandas as pd
```

```python
df = pd.read_csv('Data/fake_reg.csv')
df.head()
```

### ii. Explore the data

Lets see if there is correlation between the features and the `price` of this made up product.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(df)
```

Of this one form of visualization, we can see a strong correaltion between `feature2` and `price`

### iii. Test/Train Split

```python
from sklearn.model_selection import train_test_split
```

We need to convert the Pandas DataFrame to a Numpy array for Keras

```python
# Features
X = df[['feature1','feature2']].values

# Label
y = df['price'].values

# Split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

### iv. Normalizing/Scaling the Data

We scale the feature data. This is bc we are dealing with weights and biases in our NN. And if there are really high or low values, it could throw off our errors. 

But we do not need to scale the labels, [here's why](https://stats.stackexchange.com/questions/111467/is-it-necessary-to-scale-the-target-value-in-addition-to-scaling-features-for-re). TLDR; the feature is actually what is being passed into the NN, whereas the labels are being compared later on with the predicted outcome of the NNs. 

```python
from sklearn.preprocessing import MinMaxScaler

# To get more info on the Class
help(MinMaxScaler)
# 
# TLDR
# MinMaxScaler is dependednt of the standard dev, min, and max vlaues of the features. 
```

```python
# Creating a Scaler object from MinMaxScaler class
scaler = MinMaxScaler()
```

```python
# Fit the training data to the scaler aka train the model
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Now everything is scaled and fitted
# We can see that all the data is scaled between 0 and 1...
print(f"Train Min: {X_train.min()}")
print(f"Train Max: {X_train.max()}")
```

# The Model


### i. Imports

There are several ways you can import Keras from Tensorflow. For now we will use the method shown in the **official TF documentation**.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential

help(Sequential)
```

## I. Creating a Model

There are two ways to create models through the TF 2.0 Keras API... 
- either pass in a list of layers all at once 
- or add them one by one

For the sake of examples, I'll show both.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

# DENSE: 
#    - A reqular, densely layered NN
#       - Where every neuron is connected to every other neuron in the next layer. 
```

### i. Model - as a list of layers

```python
model = Sequential([
    Dense(units=2),
    Dense(units=2),
    Dense(units=2)
])
```

### ii. Model - adding in layers one by one

```python
model = Sequential()

model.add(Dense(2))
model.add(Dense(2))
model.add(Dense(2))
```

This will be the perfered method of this lecture. This is bc it is much more customizable. You can add and remove layers at will. So, let's build a simple model and then compile it by defining our solver...

```python
model = Sequential()

model.add(Dense(4,activation='relu')) # NOTE: 'relu' = rectified linear unit
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))

# Final output node for prediction
model.add(Dense(1))

model.compile(optimizer='rmsprop',loss='mse')
```

### iii. Choosing an optimizer and loss

<ins>_Optimizer_</ins>  
- The method of gradient decent. There are many, so look within the `help()` of the method to see whats availible. 

<ins>_Loss_</ins>  
To choose a loss, we need to keep in mind what kind of problem we are trying to solve:

    # For a multi-class classification problem
    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # For a binary classification problem
    model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # For a mean squared error regression problem
    model.compile(optimizer='rmsprop',
                  loss='mse')

## II. Training the Model

Below are some common definitions that are necessary to know and understand to correctly utilize Keras:

* Sample: 
    + one element of a dataset.
        + ex1: one image is a sample in a convolutional network
        + ex2: one audio file is a sample for a speech recognition model
* Batch: 
    + a set of N samples. 
        + The samples in a batch are processed independently, in parallel. 
     + If training, a batch results in only one update to the model. A batch generally approximates the distribution of the input data better than a single input. 
         + The larger the batch, the better the approximation; 
         + however, it is also true that the batch will take longer to process and will still result in only one update. 
             + i.e. (evaluate/predict), it is recommended to pick a batch size that is as large as you can afford without going out of memory (since larger batches will usually result in faster evaluation/prediction).
* Epoch: 
    + an arbitrary cutoff, generally defined as "one pass over the entire dataset" 
    + used to separate training into distinct phases
        + useful for logging and periodic evaluation.
    + When using `validation_data` or `validation_split` with the fit method of Keras models, evaluation will be run at the end of every epoch.
    + Within Keras, there is the ability to add callbacks specifically designed to be run at the end of an epoch.
        + Examples of these are learning rate changes and model checkpointing (saving).

```python
model.fit(X_train,y_train,epochs=250)
```

## III. Evaluation

Let's evaluate our performance on our training set and our test set. We can compare these two performances to check for overfitting.

```python
loss_df = model.history.history['loss']
```

```python
sns.lineplot(x=range(len(loss_df)),y=loss_df)
plt.title("Training Loss per Epoch");
```

So we see the losses of hte training dataset. But what we really want to see is how the model performs on on data it hasnt seen before, i.e. the test dataset. 

### i. Compare final evaluation (MSE) on training set and test set.

These should hopefully be fairly close to each other.

```python
model.metrics_names
```

```python
training_score = model.evaluate(X_train,y_train,verbose=0)
test_score = model.evaluate(X_test,y_test,verbose=0)
```

```python
training_score
```

```python
test_score
```

As we can see, the MSE (aka. the loss) of the training data and test data are very close (training mse = $23.85$ while test mse = $25.57$). 

This is a way to interprete our prediction performace results. 

### ii. Further Evaluations

Another way of evaluating prediction performace is with _root mean square error_.  

```python
# Pass in the test features so the model can predict what the price is off the features
test_predictions = model.predict(X_test)
test_predictions
```

```python
# Turn the actual results into a DF
pred_df = pd.DataFrame(y_test,columns=['Test Y'])

# Then convert the predicted results into a series
test_predictions = pd.Series(test_predictions.reshape(300,))
test_predictions
```

```python
# Concatenate the actual results with the the model's predictions
pred_df = pd.concat([pred_df,test_predictions],axis=1)
pred_df.columns = ['Test Y','Model Predictions']
pred_df
```

Let's compare to the real test labels!

```python
sns.scatterplot(x='Test Y',y='Model Predictions',data=pred_df)
# As we can see, the model is perfoming quite well, but lets confirm with the errors...
```

```python
# Add another column in the prediction df to show the errors
pred_df['Error'] = pred_df['Test Y'] - pred_df['Model Predictions']

# Then graph those errors
sns.displot(pred_df['Error'],bins=50)
```

Lets dive into ohter methods to evaluate the models prediction performance...

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
# Observation:
#    - We are, on average, off by $4
#    - The mean price is ~$500, therefore an error of $4 isnt too bad at all
# 
mean_absolute_error(pred_df['Test Y'],pred_df['Model Predictions'])
```

```python
mean_squared_error(pred_df['Test Y'],pred_df['Model Predictions'])
```

```python
# Essentially the same thing, difference just due to precision
test_score
```

```python
#RMSE
test_score**0.5
```
